- The goal of this part is to build a neural network and tweak it with different parameter values and compare the results of each model. your work will be based on Fashion MNIST dataset. This is a dataset of 60,000 28x28 grayscale images of 10 fashion categories, along with a test set of 10,000 images. The categories of the Fashion MNIST dataset are T-shirt/top, Trouser, Pullover, Dress, Coat, Sandal, Shirt, Sneaker, Bag or Ankle boote.

### Configuration 1

- HIDDEN ACTIVATION    : tanh <br />
- HIDDEN INITIALIZER   : Uniform 0.1 <br />
- OUTPUT ACTIVATION    : sigmoid <br />
- OUTPUT INITIALIZER   : Uniform 0.1 <br />
- LOSS FUNCTION        : MSE <br />
- OPTIMIZER            : SGD lr=0.01 <br />
- MINI-BATCH SIZE      : 1 <br />

```python
%matplotlib inline

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.utils import to_categorical
from matplotlib import pyplot
import numpy as np
import logging
tf.get_logger().setLevel(logging.ERROR)

EPOCHS = 20
BATCH_SIZE = 1
```

```python
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images,
                               test_labels) = fashion_mnist.load_data()

mean = np.mean(train_images)
stddev = np.std(train_images)
train_images = (train_images - mean) / stddev
test_images = (test_images - mean) / stddev

train_labels = to_categorical(train_labels, num_classes=10)
test_labels = to_categorical(test_labels, num_classes=10)

print('dimensions of train_images: ', train_images.shape)
print('dimensions of train_labels: ', train_labels.shape)
print('dimensions of test_images: ', test_images.shape)
print('dimensions of test_images: ', test_labels.shape)
```

```python
initializer = keras.initializers.RandomUniform(
    minval=-0.1, maxval=0.1)

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(25, activation='tanh',
                       kernel_initializer=initializer,
                       bias_initializer='zeros'),
    keras.layers.Dense(10, activation='sigmoid',
                       kernel_initializer=initializer,
                       bias_initializer='zeros')])
```

```python
opt = keras.optimizers.SGD(learning_rate=0.01)
model.compile(loss='mean_squared_error', optimizer = opt,metrics =['accuracy'])
history = model.fit(train_images, train_labels, validation_data=(test_images, test_labels), epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=2, shuffle=True)

```

```python
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
pyplot.show()

pyplot.plot(history.history['accuracy'], label='train')
pyplot.plot(history.history['val_accuracy'], label='test')
pyplot.legend()
pyplot.show()
```

# Part II

- The goal of this part is to build the best neural network you can, which will categorize a given image into its respective category based on Fashion MNIST dataset: T-shirt/top, Trouser, Pullover, Dress, Coat, Sandal, Shirt, Sneaker, Bag or Ankle boote. Your model should not overfit, as much as possible, while simultaneously getting the highest score it can on the test data.


  - builds a neural network, using Keras, to predict the category of the input images,
  - trains the network on the training data, and prints out the final training accuracy,
  - evaluates the network on the test data, and prints out the test accuracy.
  - creates a plot of the model's training loss as a function of epoch.
  - creates a plot of the model's training accuracy as a function of epoch.
  - creates a plot of the model's testing loss as a function of epoch.
  - creates a plot of the model's testing accuracy as a function of epoch.


```python
model_best = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu', kernel_initializer='he_normal'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(64, activation='relu', kernel_initializer='he_normal'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(10, activation='softmax', kernel_initializer='glorot_uniform')
])

model_best.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

history_best = model_best.fit(train_images, train_labels,
                              validation_data=(test_images, test_labels),
                              epochs=20, batch_size=64,
                              verbose=2, shuffle=True)

train_loss, train_acc = model_best.evaluate(train_images, train_labels, verbose=0)
test_loss, test_acc = model_best.evaluate(test_images, test_labels, verbose=0)
print('Final training accuracy:', train_acc)
print('Final test accuracy:', test_acc)

pyplot.plot(history_best.history['loss'], label='train')
pyplot.plot(history_best.history['val_loss'], label='test')
pyplot.title('Loss')
pyplot.legend()
pyplot.show()

pyplot.plot(history_best.history['accuracy'], label='train')
pyplot.plot(history_best.history['val_accuracy'], label='test')
pyplot.title('Accuracy')
pyplot.legend()
pyplot.show()
```
